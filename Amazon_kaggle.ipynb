```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

%matplotlib inline
```

```python
df = pd.read_csv('amazon.csv')
```

```python
df.drop(columns = 'rating_count', axis = 1, inplace = True)
df = df.dropna().reset_index()
df.drop(columns = ['index'], axis = 1, inplace = True)
df.isnull().sum()[:10]
```

```python
print(df['category'])
```

```python
print(df['category'].unique())
```

```python
# Fungsi untuk transformasi kategori
def transform_category(category):
    if 'Computers' in category:
        return 'Computers'
    elif 'Electronics' in category:
        return 'Electronics'
    elif 'Health' in category or 'Personal' in category:
        return 'Health & Personal'
    elif 'Musical' in category:
        return 'Musical'
    elif 'Office' in category:
        return 'Office Product'
    elif 'Toy' in category:
        return 'Toy & Game'
    elif 'Cars' in category:
        return 'Cars'
    elif 'Home' in category:
        return 'Home'
    else:
        return 'Other'

#mengaplikasikan tranformasi pada columns category
df['category'] = df['category'].apply(transform_category)
#menampilkan data
pd.DataFrame(df['category'].value_counts()).head()
```

```python
# Menghitung jumlah kemunculan setiap kategori
category_counts = df['category'].value_counts().reset_index()
category_counts.columns = ['category', 'jumlah_penjualan']
```

```python
print(category_counts)
```

```python
# Membuat diagram batang
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='jumlah_penjualan', data=category_counts, palette='viridis')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Penjualan')
plt.title('Diagram Jumlah Penjualan Berdasarkan Kategori')
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Mengonversi kolom rating ke tipe numerik (jika belum dalam tipe numerik)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Mengelompokkan berdasarkan kategori dan menghitung rata-rata rating
average_rating_per_category = df.groupby('category')['rating'].mean()

# Menampilkan hasil
print(average_rating_per_category)
```

```python
# Plotting pie chart
plt.figure(figsize=(8, 6))
patches, texts, autotexts = plt.pie(average_rating_per_category, labels=average_rating_per_category.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Rating per Category')
plt.axis('equal')  # Memastikan lingkaran menjadi bentuk lingkaran, bukan elips

# Menambahkan nilai rata-rata di sebelah diagram
labels = [f'{label}: {avg:.2f}' for label, avg in zip(average_rating_per_category.index, average_rating_per_category.values)]
plt.legend(patches, labels, loc="best")

plt.show()
```
